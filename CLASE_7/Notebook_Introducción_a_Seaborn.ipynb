# **Barplot**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#VisualizaciÃ³n que me entregue el promedio de la variable Sepal Length segÃºn el tipo de especie
sns.barplot(x="Species",y="Sepal.Length",data=iris) # mediana el parametro adicional seria estimator = np.median
```

# **Histograma**

```python
sns.displot(iris["Sepal.Length"])
```

```python
sns.histplot(iris["Sepal.Length"])
```

```python
sns.histplot(iris["Sepal.Length"], bins=20, color="orange")
```

## **KDE Plot (Kernel Desnsity Estimation Plot)**

```python
sns.kdeplot(data = iris["Sepal.Length"])
```

```python
sns.kdeplot(data = iris["Sepal.Length"], shade = True)
```

# **Boxplot**

```python
# GrÃ¡fico de cajas univariadas

sns.boxplot(y=iris["Sepal.Length"])
```

```python
sns.boxplot(y=iris["Petal.Length"])
```

```python
# GrÃ¡fica de cajas bivariadas

sns.boxplot(data = iris, x="Species", y="Sepal.Length")
```

```python
sns.boxplot(data = iris, y="Species", x="Sepal.Length", orient = "h")
```

# **Violinplot**

```python
sns.violinplot(x=iris["Sepal.Length"], color ="green")
```

```python
sns.violinplot(data=iris, x="Species", y="Sepal.Length")
```

# **Catplot**

```python
sns.catplot(data=iris, x="Species", y="Sepal.Length")
```

# **Scatter Plot**

```python
sns.scatterplot(x="Sepal.Length", y="Sepal.Width", data = iris)
```

```python
sns.scatterplot(x="Sepal.Length", y="Sepal.Width", data = iris, hue="Species")
```

# **Profundizando con GrÃ¡ficas Bivariadas**

```python
sns.jointplot(x=iris["Sepal.Length"], y=iris["Sepal.Width"], data=iris, hue="Species")
```

```python
sns.pairplot(iris, hue="Species")
```
